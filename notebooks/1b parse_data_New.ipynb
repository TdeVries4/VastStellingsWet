```python
import os
import numpy as np
```

```python
ROOT_DIR = "/home/azureuser/cloudfiles/code/Users/T.deVries4/"  # < Indien andere user pas de ROOT_DIR aan
data_path = os.path.join(ROOT_DIR, "Vaststellings_Wet/data/output/")
book = os.path.join(data_path, "OudeVastStellingsWetTXT.txt")


text = []
with open(book, "r", encoding='latin-1') as f:
    tmp_text = f.readlines()
    print(len(tmp_text))
    text.extend(tmp_text)
    print(len(text))
```

```python
import re

current_artikel = "Artikel"
collect = False
artikel_text = []
artikel_dict = {}

for line in text:
    # Check if the line starts with "Artikel", the 9th character is a digit, and the line ends with a digit or a closing bracket
    if line.startswith("Artikel") and line[8].isdigit() and (line.strip()[-1].isdigit() or line.strip()[-1] == ']'):
        collect = True
        # If the line is different from the current article
        if line != current_artikel:
            # Store the current article and its text in the dictionary
            artikel_dict[re.sub(r' \[.*?\]', '', current_artikel)] = " ".join(artikel_text).replace("\n", "").replace("\t", "")
            # Update the current article
            current_artikel = line.replace("\n", "")
            # Reset the text collection for the new article
            artikel_text = []        
            continue  # Skip adding the article number to the text
        else:
            # Update the current article
            current_artikel = line    
                   
    if collect:
        # Collect the lines of the current article
        artikel_text.append(line)

```

De code loopt door alle regels in de lijst text heen en controleert of de regel met "Artikel" begint. \n = verwijderen van ongwenste nieuwe regels \t = verwijderen van ongewenste tabs artikel_dict betekend dat het een eigen woordenboek maar waarin elk artikelnummer wordt gekoppeld aan de volledige tekst van het artikel.
